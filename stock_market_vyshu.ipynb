```python

all_data = pd.DataFrame()

for file in company_list:
    current_df = pd.read_csv(file)
    all_data = pd.concat([all_data, current_df], ignore_index=True)


```

```python
plt.figure(figsize=(23,11))
for index, company in enumerate(comp_names, 1):
    if index > 4:
        break  # Exit the loop if more than 4 subplots are needed

    plt.subplot(2, 2, index)
    filter1 = all_data['Name'] == company
    df = all_data[filter1]
    plt.plot(df['date'], df['close'])
    plt.title(company)

plt.tight_layout()
plt.show()
```

```python
closing_price=pd.DataFrame()
```

```python
closing_price['aapl_close']=aapl['close']
closing_price['aiz_close']=aiz['close']
closing_price['ajg_close']=ajg['close']
closing_price['alxn_close']=alxn['close']
closing_price['ame_close']=ame['close']
```
